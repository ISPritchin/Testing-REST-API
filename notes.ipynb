# Зачём тестировать API
* легче поддерживать, меньше требований к инфраструктуре
* раз тесты дешевле, их количество может быть больше
* API-тесты более стабильны, в отличии от UI тестов (тут не меняются локаторы и т. п.)
* Тестировать API можно и без клиента
* При помощи тестирования API можно проверить безопасность и стабильность работы сервера.
* API тесты могут ускорить UI тесты

# Стек технологий
Для автоматизации API требуется:
* приложение
* язык программирования
* тестовый фреймворк
* HTTP-библиотека
* инструмент для отчётов


# Введение в HTTP и API

Виды интерфейсов:
* API (программный интерфейс приложения)
* GUI (графический интерфейс пользователя)
* CLI (интерфейс командной строки)

API (application programming interface) -- программный интерфейс приложения. Когда тестировщик проверяет API созданное для веб-сайта или мобильного приложения, он подменяет одну из сторон общения, при этом продолжает общаться с понятным программе языке.

Протокол -- некоторый формат общения; договоренность между собеседниками, как отсылать и принимать информацию и как её понимать. Примеры протокола: сигнальные костры, азбука Морзе. В протоколе важны две вещи:
* способ передачи
* формат сигналов

HTTP - протокол прикладного уровня, он предназначен для передачи текста в определенном формате.

# Структура HTTP-запроса и HTTP-ответа
Структура HTTP запроса:
* URL
* Заголовок запроса
* Тело запроса


Структура HTTP ответа:
* Код ответа. Основные коды:
  - 200 - успешный запрос
  - 301 - перенаправление на другой URL
  - 403 - ресурс запрещен для клиента
  - 404 - запрос на пустой URL
  - 500 - сервер не обработал запрос
* Заголовки
* Тело ответа

Такой подход позволяет запрашивать и получать информацию в заданном виде.

Список запросов и параметров, которые имеют смысл для сервера и есть API! Разработчики приложения решают какие запросы, на какой адрес должны приходить для получения от сервера тех или иных действий. API может использовать разные протоколы для общения приложений (в том числе и HTTP).

HTTP API - API, использующий для общения протокол HTTP

Таким образом, API:
* способ общения одной программы с другой (например, клиента с сервером) при помощи запросов
* клиент создаёт HTTP-запросы, которые обрабатывает сервер и возвращает HTTP-ответ
* клиент может реагировать на ответ сервера
* тестирование API - проверка на то, что сервер реагирует адекватно на запросы, которые могут прийти от клиента и на запросы, которые клиент отправлять ему не должен.



# Cookie

Куки -- специальный файл, который создаётся клиентом на основе ответа сервера, имеющие срок годности, после чего удаляются клиентом. У каждого куки есть:
* имя
* значение
* домен

Куки применяются, например, в авторизации:
* При первом посещении сайта пользователя сервер запрашивает логин и пароль
* При успешной авторизации отправляются куки, которые имеются только у этого пользователя
* В следующий раз, когда клиент снова будет делать HTTP запрос на сервер, он приложит значение куки и сервер узнает пользователя
